# Data Preprocessing

## Feature Extraction

### Visitor

```python
df['visitor'] = df['adults']+df['children']+df['babies']
```

### Total_Booking

```python
df['Total_Booking'] = df['previous_cancellations']+df['previous_bookings_not_canceled']
```

```python
df[ 'previous_bookings_not_canceled'].value_counts()
```

```python
df.columns
```

## Handling Feature

### Drop Feature

```python
df = df.drop(['lead_time','adults', 'children', 'babies',
              'distribution_channel','previous_cancellations', 'previous_bookings_not_canceled'], axis = 1)
df.sample(5)
```

### Required Car Parking Spaces - Invalid Values

```python
df['required_car_parking_spaces'] = np.where(df['required_car_parking_spaces']==0, 'Not Required', 'Required')
```

### Meal - Invalid Values

```python
df['meal']=np.where(df['meal'].isin(['Breakfast']), "Breakfast",
                               np.where(df['meal'].isin(['Full Board']), 'Full Board',
                                        np.where(df['meal'].isin(['Undefined']), 'No Meal', 'Dinner')))
df.head()
```

## Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('is_canceled', axis=1),df['is_canceled'] ,
                                   random_state=42, 
                                   test_size=0.2, shuffle=False)
```

```python
train = X_train.join(y_train)
train.shape
```

```python
test = X_test.join(y_test)
test.shape
```

## Handle Outliers

### Stays in Weekend Nights > 11

```python
train.columns
```

```python
for i in ['stays_in_weekend_nights']:
  print(f'Jumlah baris sebelum memfilter',i, 'outlier:', len(train))
  a = len(train)
  train=train[train[i]<=11]
  print(f'Jumlah baris setelah memfilter',i, 'outlier:', len(train))
  b = len(train)
  print (f'persen yang dibuang : {100-((b/a)*100)}')
```

### Stays in Weekdays Nights

```python
for i in ['stays_in_weekdays_nights']:
  print(f'Jumlah baris sebelum memfilter',i, 'outlier:', len(train))
  a = len(train)
  train=train[train[i]<=35]
  print(f'Jumlah baris setelah memfilter',i, 'outlier:', len(train))
  b = len(train)
  print (f'persen yang dibuang : {100-((b/a)*100)}')
```

### Days in Waiting List

```python
for i in ['days_in_waiting_list']:
  print(f'Jumlah baris sebelum memfilter',i, 'outlier:', len(train))
  a = len(train)
  train=train[train[i]<=300]
  print(f'Jumlah baris setelah memfilter',i, 'outlier:', len(train))
  b = len(train)
  print (f'persen yang dibuang : {100-((b/a)*100)}')
```

### ADR 

```python
for i in ['adr']:
  print(f'Jumlah baris sebelum memfilter',i, 'outlier:', len(train))
  a = len(train)
  train=train[train[i]<=1000]
  print(f'Jumlah baris setelah memfilter',i, 'outlier:', len(train))
  b = len(train)
  print (f'persen yang dibuang : {100-((b/a)*100)}')
```

### Visitor

```python
for i in ['visitor']:
  print(f'Jumlah baris sebelum memfilter',i, 'outlier:', len(train))
  a = len(train)
  train=train[train[i]<=15]
  print(f'Jumlah baris setelah memfilter',i, 'outlier:', len(train))
  b = len(train)
  print (f'persen yang dibuang : {100-((b/a)*100)}')
```

### Total Booking

```python
for i in ['Total_Booking']:
  print(f'Jumlah baris sebelum memfilter',i, 'outlier:', len(train))
  a = len(train)
  train=train[train[i]<=40]
  print(f'Jumlah baris setelah memfilter',i, 'outlier:', len(train))
  b = len(train)
  print (f'persen yang dibuang : {100-((b/a)*100)}')
```

### Company

```python
for i in ['company']:
  print(f'Jumlah baris sebelum memfilter',i, 'outlier:', len(train))
  a = len(train)
  train=train[(train[i]<=190)&
             (train[i]>=170)]
  print(f'Jumlah baris setelah memfilter',i, 'outlier:', len(train))
  b = len(train)
  print (f'persen yang dibuang : {100-((b/a)*100)}')
```

### Agent

```python
for i in ['agent']:
  Q1 = train[i].quantile(.25)
  Q3 = train[i].quantile(.75)
  IQR = Q3 - Q1
  Batas_Atas = Q3 + 1.5*IQR
  Batas_Bawah = Q1 - 1.5*IQR
  print(f'Jumlah baris sebelum memfilter',i, 'outlier:', len(train))
  a = len(train)
  train=train[(train[i]<=Batas_Atas)&
              (train[i]>=Batas_Bawah)]

  print(f'Jumlah baris setelah memfilter',i, 'outlier:', len(train))
  b = len(train)
  print (f'persen yang dibuang : {100-((b/a)*100)}')
```
